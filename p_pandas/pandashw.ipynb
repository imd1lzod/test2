```python
import pandas as pd
data = pd.read_csv('adult.data.csv')
data.head()
```

```python
# 1
data['sex'].value_counts()
```

```python
# 2
print(data.loc[data['sex'] == 'Female']['age'].mean())
```

```python
# 3
germans = data.loc[data['native-country'] == 'Germany']
len(germans) * 100 / len(data)
```

```python
# 4-5
more_than = data.loc[data['salary'] == '>50K']
less_than = data.loc[data['salary'] == '<=50K']
print(more_than['age'].mean())
print(more_than['age'].std())
print(less_than['age'].mean())
print(less_than['age'].std())
```

```python
# 6
more_than = data.loc[data['salary'] == '>50K']
more_than['education'].unique()
```

```python
# 7
df = data.groupby(['age', 'sex'])
df.head()
df.describe()
data[data['race'] == 'Amer-Indian-Eskimo']['age'].max()
```

```python
# 8
more_than = data.loc[data['salary'] == '>50K']

single = ['Never-married','Divorced','Separated','Widowed']
married = ['Married-civ-spouse','Married-spouse-absent','Married-AF-spouse']

single_count = more_than[more_than['marital-status'].isin(single)].shape[0]
married_count = more_than[more_than['marital-status'].isin(married)].shape[0]

print(single_count, married_count)
```

```python
# 9
data['hours-per-week'].max()
len(data[data['hours-per-week'] == data['hours-per-week'].max()])
more_workers = data[data['hours-per-week'] == data['hours-per-week'].max()]
len(more_workers[more_workers['salary'] == '>50K']) * 100 / len(more_workers)
```

```python
# 10
more_than.groupby('native-country')['hours-per-week'].mean()
less_than.groupby('native-country')['hours-per-week'].mean()

more_than[more_than['native-country'] == 'Japan']['hours-per-week'].mean()
less_than[less_than['native-country'] == 'Japan']['hours-per-week'].mean()
```
