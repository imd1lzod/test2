```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {
        "name": [" Ali", "vali ", "SAMI", "Ali"],
        "salary": [2000, None, 3000, 2000]
    }
)

df.to_csv('advanced.csv', index=False)
data = pd.read_csv('advanced.csv')


data['name'] = data['name'].str.strip().str.lower()
data['salary'] = data['salary'].fillna(data['salary'].mean())
data = data.drop_duplicates(subset=['name', 'salary'])
data
```
